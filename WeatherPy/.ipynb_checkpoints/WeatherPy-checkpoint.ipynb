```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from pprint import pprint
import os

# Import openweathermap API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy as cp

# Output File (CSV)
output_data_file = os.path.join("..","output_data","cities.csv")

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
city_names = []
city_countries = []

# Create a set of random lat and lng combinations. Because the earth is mostly coovered in water, most of
# these coordinates will be in the oceans.
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = cp.nearest_city(lat_lng[0], lat_lng[1])
    
    # If the city is unique, then add it to a our cities list.
    if city not in cities:
        cities.append(city)
        city_names.append(city.city_name)
        city_countries.append(city.country_code)
```

```python
# Print the city count to confirm sufficient count
len(city_names)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# here the base url is for current weather data. Using historical data would be better for this exercise, 
# but it aint free
current_weather_base_url = "http://api.openweathermap.org/data/2.5/weather?"
```

```python
# initialize the dataframe with city names and their countries. weather information will start as Nones
cities_df = pd.DataFrame({"Name": city_names, "Country Code": city_countries})
cities_df["Latitude"] = None
cities_df["Longitude"] = None
cities_df["Temp"] = None
cities_df["Humidity"] = None
cities_df["Cloudiness"] = None
cities_df["Wind Speed"] = None
cities_df.head()
```

```python
weather_params = \
{
    "q": "cherskiy" + "," + "ru",
    "units": "imperial",
    "appid": api_key
}
response = requests.get(current_weather_base_url, params=weather_params)
response_json = response.json()
```

```python
# iterating over the rows of the cities dataframe
for index, row in cities_df.iterrows():
    # creating the parameters for each API query for each city
    weather_params = \
    {
        "q": row["Name"] + "," + row["Country Code"],
        "units": "imperial",
        "appid": api_key
    }
    print(f"Looking up weather data for record {index}--{row['Name'] + ',' + row['Country Code']}:")
    # query. supposedly openweathermap is limited to 60 requests per minute, but the requests module seems
    # to handle overages on its own. 
    # again, historical data would be better, but it's not available from this api for free.
    response = requests.get(current_weather_base_url, params=weather_params)
    response_json = response.json()
    
    # only filling out the dataframe if the response is successful
    if response.status_code == 200:
        print(f"City found! Adding weather data for {row['Name'] + ',' + row['Country Code']}.")
        row["Latitude"] = response_json["coord"]["lat"]
        row["Longitude"] = response_json["coord"]["lon"]
        row["Temp"] = response_json["main"]["temp"]
        row["Humidity"] = response_json["main"]["humidity"]
        row["Wind Speed"] = response_json["wind"]["speed"]
        row["Cloudiness"] = response_json["clouds"]["all"]
    else:
        print(f"No data found for {row['Name'] + ',' + row['Country Code']}.")
```

```python
# if the response was unsuccessful, no data was found for that city, and it needs to be dropped from the
# dataframe before plotting
cities_df.dropna(inplace=True)
cities_df.to_csv(output_data_file)

cities_df.shape
```

#### Latitude vs. Temperature Plot

```python
plt.figure(figsize=(15,10), facecolor = "w")
ax = plt.axes()
ax.set_axisbelow(True)
ax.set_facecolor("xkcd:light grey")
plt.grid(axis="both", color = "w")

plt.scatter(cities_df["Latitude"], cities_df["Temp"], alpha=0.75)

plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.title("Temperature vs. Latitude (11/16/2019)", size = "xx-large")
plt.annotate('Strong relationship between temperature and latitude. Temperature peaks at 0 Latitude. ' + \
             'Far more cities with positive latitude (Northern Hemisphere) than negative \n(Southern ' + \
             'Hemisphere), which comports with the fact that the majority of the population is in ' + \
             'the north.', 
             (0,0), (0, -50), xycoords='axes fraction', textcoords='offset points', va='top')

plt.savefig(os.path.join("..","Images","Temperature vs Latitude (11-16-2019).png"), bbox_inches="tight")
```
